```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/Users/hasanismayilov/My Git Projects/Portfolio/my-exams/supermarket_sales/supermarket_sales - Sheet1.csv')
df.head()
```

✅ Question 1 – Bar Plot: Total Sales by Product Line
Task:
Create a bar chart that shows the total sales (Total) for each Product line.

Instructions:

Use Seaborn for plotting.
X-axis: Product line
Y-axis: Sum of Total
Add titles and rotate labels if needed for readability.

```python
sns.barplot(data=df, x='Product line', y='Total', estimator=sum, errorbar=None)
plt.xlabel('Product line')
plt.ylabel('Total')
plt.title('Total by Product line')
plt.xticks(rotation=45)
plt.show()
```

✅ Question 2 – Count Plot: Gender Distribution

🟦 Question 5 – Histogram: Unit Price Distribution
Task:
Create a histogram to visualize the distribution of Unit price across all sales.

✅ Question 6 – Violin Plot: Total Sales by Customer Type

🟩 Question 7 – Line Plot: Average Rating Over Time
Task:
Plot the average of Rating for each Date to visualize how customer satisfaction trends over time.

Exact columns to use:

Date → x-axis
Rating → y-axis (averaged per day)
📌 Note: You'll need to convert 'Date' to datetime first.

```python
df['Date'] = pd.to_datetime(df['Date'])
grouped_df = df.groupby('Date')['Rating'].agg('mean').reset_index()
sns.lineplot(data = grouped_df, x = 'Date', y = 'Rating')
plt.title('Average Rating Over Time')
plt.xlabel('Date')
plt.ylabel('Average Rating')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

✅ Question 8 – Strip Plot: Rating by Branch
Task:
Create a strip plot to visualize how Rating values are distributed across different **Branch**es.

Exact columns to use:

Branch → x-axis
Rating → y-axis

```python
plt.figure(figsize = (8,12))
sns.stripplot(data=df, x='Branch', y='Rating', jitter=True, palette='Set2')
plt.title('Rating Distribution')
plt.xlabel('Branch')
plt.ylabel('Distribution')
plt.tight_layout()
plt.show()
```

🧠 Question 9 – Visualization (Intermediate): Correlation Heatmap
Task:
Visualize the correlation between numeric features in the dataset using a heatmap.

Exact instruction:

Use all numeric columns like 'Unit price', 'Quantity', 'Tax 5%', 'Total', 'cogs', 'gross income', 'Rating', etc.
Create a Seaborn heatmap to show how strongly they’re correlated.
Add a title and color scale.

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize = (8,8))
sns.heatmap(numeric_df.corr(), cmap='plasma', annot=True, fmt=".2f")
plt.show()
```

✅ Question 10 – Pie Chart: Payment Method Distribution
Task:
Create a pie chart to show the proportion of each Payment method used by customers.

Exact column to use:

'Payment'

```python
payment_count = df['Payment'].value_counts().reset_index()
payment_count.columns = ['payment', 'count']
plt.pie(payment_count['count'], labels = payment_count['payment'], autopct='%1.1f%%', startangle=140)
plt.title('Payment Method Share')
plt.tight_layout()
plt.show()
```

✅ Question 11 – Grouped Bar Plot: Average Gross Income by Branch and Gender

✅ Question 13 – Horizontal Bar Chart: Average Rating by Customer Type
Task:
Create a horizontal bar chart that shows the average Rating for each Customer type.

Exact columns to use:

'Customer type' → y-axis (since it’s horizontal)
'Rating' → x-axis (average)

```python
plt.figure(figsize = (6,3))
sns.barplot(data=df, x = 'Rating', 
            y = 'Customer type', 
            estimator='mean', 
            errorbar= ('ci', 95),  
            orient = 'h', palette= 
            'Set2')
plt.title('Average Rating by Customer Type', loc = 'center', pad=0.3)
plt.xlabel('Rating')
plt.ylabel('Customer Type')
plt.tight_layout()
plt.show()
```

✅ Question 14 – Line Plot: Daily Gross Income Over Time
Task:
Create a line chart showing total gross income per Date to observe trends over time.

Exact columns to use:

'Date' → x-axis (make sure it's datetime)
'gross income' → y-axis (sum per day)

```python
df['Date'] = pd.to_datetime(df['Date'])
grouped_df = df.groupby('Date')['gross income'].sum().reset_index()
plt.figure(figsize = (8, 4))
sns.lineplot(data = grouped_df, 
             x = 'Date',
             y = 'gross income',
             orient = 'x', 
             markers= 'o')
plt.title('Total Gross Income Over Time')
plt.tight_layout()
plt.show()
```

✅ Question 15 – Box Plot: Unit Price Distribution by Product Line
Task:
Create a box plot showing the distribution of Unit price for each Product line.

Exact columns to use:

'Product line' → x-axis
'Unit price' → y-axis
Rotate labels for clarity, and style it clean.

✅ Question 16 – Bar Chart: Total Sales by Payment Method
Task:
Create a bar chart showing the total Total value (sales amount) for each Payment method.

Exact columns to use:

'Payment' → x-axis
'Total' → y-axis (sum per payment method)

```python
df_g = df.groupby('Payment')['Total'].sum().reset_index()
df_g['Payment'] = df_g['Payment'].replace({
    'Cash': 'Dollar',
    'Credit card': 'Card',
    'Ewallet': 'Electron'
})
sns.set_theme('notebook')
sns.set_style('darkgrid')
plt.figure(figsize = (6, 6))
sns.barplot(data = df_g, x = 'Payment', y = 'Total', palette= 'Set2')
plt.title('Total by Payment method')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

✅ Question 20 – Heatmap: Average Rating by Branch and Gender
Task:
Create a heatmap that shows the average Rating for each combination of Branch and Gender.

Exact columns to use:

'Branch' → rows
'Gender' → columns
'Rating' → values (averaged)

```python
pivoted_df = df.pivot_table(index='Branch', columns='Gender', values='Rating', aggfunc = 'mean')
sns.heatmap(data=pivoted_df, cmap = 'coolwarm', annot=True)
plt.tight_layout()
plt.show()
```
