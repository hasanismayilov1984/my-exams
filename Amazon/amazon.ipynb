✅ Q10: “Sentiment Looping”
Loop through the review_content column.
Create a new column called sentiment_tag with these rules:


If the review contains...	Set sentiment_tag to:
Contains word "great"	'positive'
Contains word "bad"	'negative'
Contains both "great" & "bad"	'mixed'
Otherwise	'neutral'

```python
from numpy import append
import pandas as pd


sentiment_tag = []
for text in df['review_content']:
    if isinstance(text, str):
        words = text.lower()
        if 'great' in words and 'bad' in words:
            sentiment_tag.append('mixed')
        elif 'bad' in words:
            sentiment_tag.append('negative')
        elif 'great' in words:
            sentiment_tag.append('positive')
        else: sentiment_tag.append('neutral')
    else: sentiment_tag.append('no data')
df['sentiment_tag'] = sentiment_tag
df['sentiment_tag'].value_counts()
```

✅ Q11 – “Apply Like a Loop Lord”
Create a column called word_flags using .apply() with a lambda or function.
For each row in review_content:
Loop through the words in the text
Build a list of:
'🔥' if the word is 'great'
'💩' if the word is 'bad'
Ignore case
Return the list of flags for each review

```python
def word_function(text):
    tag_flag = []
    if isinstance(text, str):
        for word in text.lower().split():
            word = word.strip(',.!?"\'')
            if word == 'great':
                tag_flag.append('great')
            elif word == 'bad':
                tag_flag.append('bad')
    else: tag_flag.append('noting there bro')            
    return tag_flag if tag_flag else ['neural']
df['word_flag'] = df['review_content'].apply(word_function)
df['word_flag'].value_counts().head()
```

✅ Q12 (Your Turn)
Create two new columns:

great_count — how many times "great" appears in each review_content
bad_count — how many times "bad" appears
📌 Rules:

Case-insensitive ("Great" = "great")
Strip punctuation like . or ,
Use .apply() with a custom function
You return both counts → assign to two new columns

```python
def mood_count(text):
    bad = 0
    great = 0
    if isinstance(text, str):
        for word in text.lower().split():
            word = word.strip(',.?!')
            if word == 'bad':
                bad += 1
            elif word=='great':
                great += 1
    return pd.Series([bad, great])
df[['bad_count', 'great_count']] = df['review_content'].apply(mood_count)
df.head()
```

✅ Bonus 3: Build new scores based on word length!

🧠 Plan:
Build a custom score for the review:
+1 for each 'short'
+2 for each 'medium'
+5 for each 'long'
Total all points for each review
Create a new column like:
df['total_word_score']

```python
def custom_score(text):
    short = 0
    medium = 0
    long = 0
    if isinstance(text, list):
        for word in text:
            if word == 'short':
                short += 1
            elif word == 'medium':
                medium += 1
            else:
                long += 1
    return (short*1 + medium*2 + long*3)
df['total_word_score'] = df['word_length'].apply(custom_score)
df['total_word_score']
```
